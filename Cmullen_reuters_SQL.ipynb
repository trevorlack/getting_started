```python
import pandas as pd
import os

os.chdir('S:/Trevor/Quant Projects/Corporate Bond Liquidity')
df = pd.read_csv('Eikon_Bond_DataHC.csv', index_col=0)
df
```

```python
dictionary = df.to_dict()
dictionary
```

```python
reform = {(outerKey, innerKey): values for outerKey, innerDict in dictionary.items() for innerKey, values in innerDict.items()}
tup = list(reform.keys())
reform

```

```python
df2 = pd.DataFrame.from_dict(reform, orient='index')
df2
```

```python
df2.columns=['Par']
df2.index = pd.MultiIndex.from_tuples(tup, names=['Cusip', 'Date'])
df2
```
